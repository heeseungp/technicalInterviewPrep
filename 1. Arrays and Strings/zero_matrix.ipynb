Zero Matrix: Write an algorithm such that if an element in an MxN matrix is 0, it's entire row and column are set to 0.

```python
'''
Space complexity: O(Row + Col)
Time complexity: O(Row * Col)
'''
def nullify(row, col, usrinput):
    for i in range(0,len(row)):
        if row[i] == True:
            for j in range(0,len(usrinput[0])):
                usrinput[i][j] = 0
    
    for i in range(0,len(col)):
        if col[i] == True:
            for j in range(0,len(usrinput)):
                usrinput[j][i] = 0
    
    return usrinput

def zero_matrix(usrinput):
    if len(usrinput) > 0 and len(usrinput[0]) > 0:
        row = [False] * len(usrinput)
        col = [False] * len(usrinput[0])
        for i in range(0,len(usrinput)):
            for j in range(0,len(usrinput[0])):
                if usrinput[i][j] == 0:
                    row[i] = True
                    col[i] = True
        usrinput = nullify(row,col,usrinput)
        return usrinput

'''
Better solution
Space complexity: O(1)
Time Complexity: O(Row * Col)
'''
def optimal_zero_matrix(usrinput):
    firstrow = False
    firstcol = False
    if len(usrinput) > 0 and len(usrinput[0]) > 0:
        for i in range(0,len(usrinput[0])):
            if usrinput[0][i] == 0:
                firstrow = True
                break
        for j in range(0,len(usrinput)):
            if usrinput[j][0] == 0:
                firstcol = True
                break
        for i in range(1,len(usrinput)):
            for j in range(1,len(usrinput[0])):
                if usrinput[i][j] == 0:
                    usrinput[i][0] = 0
                    usrinput[0][j] = 0
        
        return optimal_nullify(usrinput,firstrow,firstcol)

def optimal_nullify(usrinput, firstrow, firstcol):
    for i in range(1,len(usrinput)):
        if usrinput[i][0] == 0:
            usrinput = nullifyRow(usrinput, i)
    for j in range(1,len(usrinput[0])):
        if usrinput[0][j] == 0:
            usrinput = nullifyCol(usrinput, j)
    
    if firstrow is True:
        usrinput = nullifyRow(usrinput, 0)
    if firstcol is True:
        usrinput = nullifyCol(usrinput, 0)
    
    return usrinput


def nullifyRow(usrinput, whichrow):
    for i in range(0, len(usrinput[0])):
        usrinput[whichrow][i] = 0
    return usrinput

def nullifyCol(usrinput, whichcol):
    for i in range(0, len(usrinput)):
        usrinput[i][whichcol] = 0
    return usrinput
        
```
